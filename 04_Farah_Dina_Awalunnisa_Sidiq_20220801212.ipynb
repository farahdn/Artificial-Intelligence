```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'data.csv'  # Ganti dengan path yang sesuai jika diperlukan
df = pd.read_csv(file_path)

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Diagram Distribusi Diagnosis
# Count the occurrences of each diagnosis
diagnosis_counts = df['diagnosis'].value_counts()

# Create a bar plot for the diagnosis distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=diagnosis_counts.index, y=diagnosis_counts.values, palette="pastel")
plt.title('Distribution of Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# 2. Diagram Scatter untuk radius_mean vs texture_mean
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='radius_mean', y='texture_mean', hue='diagnosis', style='diagnosis', palette='deep', alpha=0.7)
plt.title('Scatter Plot of Radius Mean vs Texture Mean')
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.legend(title='Diagnosis')
plt.show()
```
